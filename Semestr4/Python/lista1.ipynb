## Zadanie 2 _(5pt)_

Napisz funkcję `primes(n)` zwracająca listę liczb pierwszych nie większych niż $n$ np.

```
In [ ]:   primes(6)
Out [ ]: [2, 3, 5]
```

````python
import math

def primes(n):
    tmp=[]
    for x in range(2,n+1):
        prime= True
        for y in range(2,int(math.sqrt(x))+1):
            if(x % y == 0):
                prime=False
                break
        if(prime==True):
            tmp.append(x)
    return tmp

primes(4)
````

Wykorzystaj słownik z kluczami zawierającymi indeksy znaków różnych od `'*'` i wartościami liter z wzorca. Napisz program, który znajduje wszystkie sekwencje z listy, które pasuje do podanego wzorca.

````python
def check(list , wzor):
    tmp=[]
    for slowo in list:
        same=True
        for x in range(0,len(wzor)):
            if( slowo[x]!='*'  and slowo[x] != wzor[x]):
                same=False
                break
        if( same == True):
            tmp.append(slowo)
    return tmp

````

## Zadanie 8 _(10pt)_

Napisz program, który konwerteruje liczby rzymskie na liczby arabskie. Do zapisu liczb rzymskich używa
się 7 liter: `I=1`, `V=5`, `X=10`, `L=50`, `C=100`, `D=500`, `M=1000`. Opis algorytmu zamiany z i na system rzymski można znaleźć np. [tutaj](http://www.algorytm.org/algorytmy-arytmetyczne/zamiana-z-i-na-system-rzymski.html)

````python
arabic=[1000,500,100,50,10,5,1]
roman=['M','D','C','L','X','V','I']
roman_n=7

def roman2arabic(number):
    i=0
    j=0
    result=0
    length=len(number)
    while ((j<length) and (i<roman_n)):
        if(number[j]==roman[i]):
            result+=arabic[i]
            j+=1
        elif((i % 2 == 0) and (i<roman_n-2) and (j<length-1) and ( number[j] == roman[i+2]) and (number[j+1] == roman[i])):
            result+=arabic[i]-arabic[i+2]
            j+=2
            i+=1
        elif((i % 2 == 1) and (i<roman_n-1) and (j<length-1) and ( number[j] == roman[i+1]) and (number[j+1] == roman[i])):
            result+=arabic[i]-arabic[i+1]
            j+=2
            i+=1
        else:
            i+=1
        
    if( i == roman_n):
        result=-1
        
    return result

roman2arabic("")
````
